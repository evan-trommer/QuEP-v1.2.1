# FUNCTIONS

#### The theorized approximation for the band-width $\sigma$ as a function of screen distance $x$, initial band height $y_0$ and initial band width $\delta y$ is 
$$ \sigma_y(x,y_0,\delta y)\approx \delta y\left|1-\frac{x}{f(y_0)} \right| $$
#### where $f(y_0)$ is the focal length as calculated by Marisa's formula. 

#### The theorized approximation for the distance at which the minimum band-width is achieved, denoted as $f_{band}$, as a function of initial band height $y_0$ and initial band width $\delta y$ is 
$$ f_{band}\approx f(y_0)\left(1+y_0\frac{f'(y_0)}{f(y_0)} \right) $$
#### where 
$$ f'(y_0) = \frac{p^2}{2m\gamma kr_b}\left(1-\left(\frac{y_0}{r_b}\right)^2 \right)^{-3/2}\frac{y_0}{r_b^2} = \frac{f(y_0)y_0}{r_b^2(1-(y_0/r_b)^2)} $$

```python
def y_thickness(x, y0, dy): 
    sig_y = abs((((y0-dy)/(focal(y0-dy)/6)) - (y0/(focal(y0)/6)))*x +dy)
    return sig_y

def derivative(x, y0, dy): 
    thick_derivative = (((((y0-dy)/(focal(y0-dy)/6)) - (y0/(focal(y0)/6)))*x +dy)/(y_thickness))*((y0-dy)/(focal(y0-dy)/6)) - (y0/(focal(y0)/6))
    return thick_derivative

def find_intersect1(y0, dy): 
    xint = focal(y0-dy)*(1+(y0*(1+(y0/rb)**2)/rb**2))/6
    return xint 

def y_thickness2(x,y0,dy,f1,f2):
    sig_y = abs((((y0-dy)/f1) - (y0/f2))*x +dy)
    return sig_y

def y_thickness3(x,dy,f1):
    sig_y = dy*abs(1-x/f1)
    return sig_y

def find_intersect3(y0,dy):
    xmin = focal(y0-dy)*(1+(((y0-dy)/rb)**2)*(1+((y0-dy)/rb)**2))/6
    return xmin
```
